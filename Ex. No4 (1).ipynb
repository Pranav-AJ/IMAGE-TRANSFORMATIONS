To perform image transformation such as Translation, Scaling, Shearing, Reflection, Rotation and Cropping using OpenCV and Python.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('tree.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib

# 1. Translation
rows, cols, _ = image.shape
M_translate = np.float32([[1, 0, 50], [0, 1, 100]])  # Translate by (50, 100) pixels
translated_image = cv2.warpAffine(image_rgb, M_translate, (cols, rows))

# 2. Scaling
scaled_image = cv2.resize(image_rgb, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)  # Scale by 1.5x

# 3. Shearing
M_shear = np.float32([[1, 0.5, 0], [0.5, 1, 0]])  # Shear with factor 0.5
sheared_image = cv2.warpAffine(image_rgb, M_shear, (int(cols * 1.5), int(rows * 1.5)))

# 4. Reflection (Flip)
reflected_image = cv2.flip(image_rgb, 1)  # Horizontal reflection (flip along y-axis)

# 5. Rotation
M_rotate = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)  # Rotate by 45 degrees
rotated_image = cv2.warpAffine(image_rgb, M_rotate, (cols, rows))

# 6. Cropping
cropped_image = image_rgb[50:300, 100:400]  # Crop a portion of the image

# Plot the original and transformed images
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(translated_image)
plt.title("Translated Image")
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(scaled_image)
plt.title("Scaled Image")
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(sheared_image)
plt.title("Sheared Image")
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(reflected_image)
plt.title("Reflected Image")
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(rotated_image)
plt.title("Rotated Image")
plt.axis('off')

plt.tight_layout()
plt.show()

# Plot cropped image separately as its aspect ratio may be different
plt.figure(figsize=(4, 4))
plt.imshow(cropped_image)
plt.title("Cropped Image")
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('HappyFish.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for matplotlib

# 1. Translation (shift the image by 50 pixels in x and 100 pixels in y)
M_translate = cv2.getRotationMatrix2D((0, 0), 0, 1)  # Identity transformation
M_translate[0, 2] = 50  # x-axis translation
M_translate[1, 2] = 100  # y-axis translation
translated_image = cv2.warpAffine(image_rgb, M_translate, (image.shape[1], image.shape[0]))  # Use image.shape for size

# 2. Scaling (scale the image by 1.5x along both axes)
scaled_image = cv2.resize(image_rgb, (0, 0), fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)

# 3. Shearing (apply horizontal and vertical shearing)
M_shear = cv2.getRotationMatrix2D((0, 0), 0, 1)  # Identity transformation
M_shear[0, 1] = 0.5  # Apply shear factor along x-axis
M_shear[1, 0] = 0.5  # Apply shear factor along y-axis
sheared_image = cv2.warpAffine(image_rgb, M_shear, (int(image.shape[1] * 1.5), int(image.shape[0] * 1.5)))  # Use shape for size

# 4. Reflection (flip the image horizontally)
reflected_image = cv2.flip(image_rgb, 1)  # Flip horizontally (along y-axis)

# 5. Rotation (rotate the image by 45 degrees around its center)
M_rotate = cv2.getRotationMatrix2D((image.shape[1] / 2, image.shape[0] / 2), 45, 1)  # Rotate by 45 degrees
rotated_image = cv2.warpAffine(image_rgb, M_rotate, (image.shape[1], image.shape[0]))  # Use shape for size

# 6. Cropping (crop a region from the image)
cropped_image = image_rgb[50:300, 100:400]  # Cropping a region from (x=100, y=50) to (x=400, y=300)

# Plot the original and transformed images
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(translated_image)
plt.title("Translated Image")
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(scaled_image)
plt.title("Scaled Image")
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(sheared_image)
plt.title("Sheared Image")
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(reflected_image)
plt.title("Reflected Image")
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(rotated_image)
plt.title("Rotated Image")
plt.axis('off')

plt.tight_layout()
plt.show()

# Plot cropped image separately
plt.figure(figsize=(4, 4))
plt.imshow(cropped_image)
plt.title("Cropped Image")
plt.axis('off')
plt.show()
```
